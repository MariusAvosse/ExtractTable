# Quelques exemples avec Tesseract

## 1. Extraire du texte simple à partir d'une image:

```python
#installer pytesseract:

pip install pytesseract
```

```python
# étant donné une image image1, nous allons extraire les textes qui s'y trouvent:

try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract

print(pytesseract.image_to_string(Image.open('/home/marius/Documents/image1.png')))
```

```python
#Nous pouvons récupérer les coordonnées des éléments textuels de l’image :

print(pytesseract.image_to_data(Image.open('/home/marius/Documents/image1.png')))
```

```python
#Nous pouvons récupérer les textes d'une image(image2) contenant un tableau:

print(pytesseract.image_to_string(Image.open('/home/marius/Documents/image2.png')))
```

```python
#Nous voyons que le contenant du tableau précédent a été déformé.
#Voici comment était l'image du tableau:

Image.open('/home/marius/Documents/image2.png')
```

```python
#Prenons l'exemple d'une autre image:

print(pytesseract.image_to_string(Image.open('/home/marius/Documents/image3.png')))
```

```python
#Sur l'exemple précédent, certaines informations sont coupées. Voici l'image du tableau:

Image.open('/home/marius/Documents/image3.png')
```
